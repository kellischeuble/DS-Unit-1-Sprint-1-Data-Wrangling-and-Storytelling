```python
# have a lot of things that might give us trouble reading the CSV file
# into a dataframe

df = pd.read_csv('LoanStats_2018Q4.csv')
print(df.shape)
df.head()
```

```python
# note that all of these values are bolded.. because of the extra rows
# at the top, it is reading all of the values as column headers 
# extra rows. This is seen by looking at the shape of the data

# make the header the 2nd row:
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows =1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

```python
# now we wnat to look at the NaN values in each column.. 
# last two rows have been read in as values

pd.set_option('display.max_rows', 200)
df.isnull().sum().sort_values()
```

```python
# drop some columns that are all NaN values
# have to set the axis=1 because the default for drop() 
# is to drop rows. We have to specify that we want to drop 
# columns

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
df.head()
```

```python
# cleaning up the int_rate column
# column has % signs
# values are of data type strings, probably want them to be numeric
# there is a leading space

print(df['int_rate'][5])
type(df['int_rate'][5])
```

```python
int_rate = df['int_rate'][5]

float(int_rate.strip().strip('%'))
```

```python
# We will want to do this again, so we should make a function

def int_rate_to_flt(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# Now add the column to the dataframe using the .apply() function
# Save it over previous column

df['int_rate'] = df['int_rate'].apply(int_rate_to_flt)
df.head()
```

```python
# now clean up 'emp_title' column
# most value counts are NaN
# There are a lot of duplicate values because of 
# capitalization and what not
df['emp_title'].value_counts(dropna=False)[:20]
```

```python
# see how many value counts there are total

len(df['emp_title'].value_counts())
```

```python
# unique value counts

len(df['emp_title'].unique())
```

```python
# see how many null values we have

df['emp_title'].isnull().sum()
```

```python
# clean up by making each job title capitalized
# have some sort of string value for NaN

def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.strip().title()
  else: 
    return "unknown"

# check to see if it will work on a series before chaning dataframe
examples = examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

[clean_title(example) for example in examples]
```

```python
# now apply to column
df['emp_title'] = df['emp_title'].apply(clean_title)
df.head()
```

```python
# see if this changed the number of unique values
# old length of the array containing unique values was 43000+

len(df['emp_title'].unique())
```

```python
# look for null values

df['emp_title'].isnull().sum()
```

Convert term column from string to integer


```python
df['term'].head()
```

```python
# function to strip value of "months" and whitespace.

def to_int(term_cell):
  return int(term_cell.strip("months").strip())

df['term'] = df['term'].apply(to_int)
```

```python
df['term'].head()
```

```python
# Make a column named loan_status_is_great. It should contain the integer 1 if 
# loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

df['loan_status']
```

```python
# look at specific value

df['loan_status'][2]
```

```python
# make function to check loan status:

def check_loan_status(loan_status):
  if loan_status == 'Curent' or 'Fully Paid':
    return 1
  else:
    return 0
```

```python
# apply function to loan status column
df['loan_status_is_great'] = df['loan_status'].apply(check_loan_status)
```

```python
df['loan_status_is_great']
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns.

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'])
```

```python
# Add new columns 'last_pymnt_d_month' and 'last_pymnt_d_year'
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year

df.head()
```
